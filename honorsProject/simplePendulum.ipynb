```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def simplePendulum(m, L, theta0, w0, h, t_max, g=9.81):
    
    def f1(t, theta, w):
            return w

    def f2(t, theta, w):
            return -g/L*np.sin(theta)

    Nsteps = t_max/h
    t_range = h*np.arange(Nsteps, dtype=np.float64)
    theta = np.zeros_like(t_range)
    w = np.zeros_like(t_range)
    y = np.array([theta0, w0], dtype=np.float64)
    
    A0 = np.zeros(2)
    A1 = np.zeros(2)
    A2 = np.zeros(2)
    A3 = np.zeros(2)
    
    for i, t in enumerate(t_range):
        theta[i] = y[0]
        w[i] = y[1]

        k0 = h*f1(t, y[0], y[1])
        d0 = h*f2(t, y[0], y[1])
        A0[:] = k0, d0

        k1 = h*f1(t + 0.5*h, y[0] + 0.5*k0, y[1] + 0.5*d0)
        d1 = h*f2(t + 0.5*h, y[0] + 0.5*k0, y[1] + 0.5*d0)
        A1[:] = k1, d1

        k2 = h*f1(t + 0.5*h, y[0] + 0.5*k1, y[1] + 0.5*d1)
        d2 = h*f2(t + 0.5*h, y[0] + 0.5*k1, y[1] + 0.5*d1)
        A2[:] = k2, d2

        k3 = h*f1(t + h, y[0] + k2, y[1] + d2)
        d3 = h*f2(t + h, y[0] + k2, y[1] + d2)
        A3[:] = k3, d3

        y[:] = y + (A0 + 2*A1 + 2*A2 + A3)/6
    return t_range, theta, w, L, m, g
```

```python
data = simplePendulum(0.5, 0.5, 0.5, 0, .01, 15)
```

```python
t = data[0]
theta = data[1]
w = data[2]
L = data[3]
m = data[4]
g = data[5]

x1 = L*np.sin(theta)
y1 = -L*np.cos(theta)
v = L*w
```

```python
ax = plt.subplot(1,1,1)
ax.plot(x1, y1)
ax.set_aspect(1)
```

```python
T = 0.5*m*v**2
V = m*g*y1
E = T + V
```

```python
ax = plt.subplot(3,1,1)
ax.plot(t, V)
ax.plot(t, T)
ax.plot(t, E)
```
